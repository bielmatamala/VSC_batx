## Escriure en un fitxer  

Per defecte, la funció **`open()`** només permet **llegir** un fitxer.  

Si volem **sobreescriure-hi**, hem de passar el paràmetre `'w'` (només escriptura) o `'w+'` (permet lectura + escriptura) com a segon argument. 

Exemple 1:
```python
myfile = open('test.txt', 'w')
myfile.write("Aquest text sobreescriurà qualsevol contingut anterior.")
myfile.close()
```

Exemple 2:

```python
myfile = open('test.txt', 'w+')
myfile.write("Aquest text també sobreescriurà qualsevol contingut anterior.")
myfile.close()
```

````python
# Afegeix un segon argument a la funció: 'w', que vol dir "escriure" (write).
# Si fem servir 'w+', podem tant escriure com llegir del fitxer.

my_file = open('test.txt', 'w+')
````

### ⚠️ **Atenció!**  

Obrir un fitxer amb els modes **`'w'`** o **`'w+'`** **trunca** el fitxer original,  
és a dir, **s’esborra tot el contingut anterior** abans d’escriure-hi de nou.  

👉 Un cop obert amb `'w'`, **no podràs recuperar el contingut antic**.

````python
# Escriure en el nou fitxer
my_file.write('Això es una nova línia')
````

````python
# Llegir-lo
my_file.seek(0)
my_file.read()
````

````python
my_file.close()  # Recorda de tancar sempre
````

## Afegir contingut a un fitxer  

Si passem l’argument **`'a'`**, el fitxer s’obre i el **cursor s’ubica al final** del contingut existent,  
de manera que **qualsevol text nou s’afegeix (append)** sense esborrar el que ja hi havia.

Igual que `'w+'`, el mode **`'a+'`** permet **llegir i escriure** en el fitxer.  

A més, si el fitxer **no existeix**, Python **en crearà un de nou automàticament**.

```python
myfile = open('test.txt', 'a+')
myfile.write("\nAquesta línia s’afegeix al final del fitxer.")
myfile.close()


````python
my_file = open('test.txt', 'a+')
my_file.write("\nAquesta línia s’afegeix al final del fitxer.")
````

````python
my_file.seek(0)
print(my_file.read())
````

````python
my_file.close()
````

### Afegir contingut amb `%%writefile`  

Podem fer el mateix utilitzant la **màgia de cel·la d’IPython** `%%writefile`.  
Si afegim l’opció `-a`, el contingut s’**afegeix** al final del fitxer en lloc de sobreescriure’l.

````python
%%writefile -a test.txt

Aquesta línia s'ha afegit utilitzant la màgia d'IPython.
Afegim una altra linia aquí
````

Afegeix un **espai en blanc al principi** si vols que la primera línia comenci en una línia nova, ja que Jupyter **no reconeix seqüències d’escapament** com `\n` dins de la màgia `%%writefile`.

## Iterar a través d’un fitxer  

Fem una petita demostració de com funciona un **bucle `for`** iterant sobre un fitxer de text.  
Per començar, crearem un **nou fitxer de text** utilitzant la màgia d’IPython:

````python
%%writefile nova_prova.txt
Primera línia
Segona línia
Tercera línia
````

Ara podem usar un for per iterar-nos en el fitxer:

````python
nou_fitxer = open("nova_prova.txt")
for line in open('nova_prova.txt'):
    print(line)
````

O també la funció **readlines** que hem vist anteriorment:

````python
nou_fitxer.seek(0)
llista_linies = nou_fitxer.readlines()
print(llista_linies)
````
