# Bucles **while**

La instrucció **`while`** en Python és una de les formes més generals de realitzar **iteracions**.  
Una instrucció **`while`** executa repetidament una o diverses línies de codi **mentre la condició sigui certa**.  

Se l’anomena un *bucle* perquè les instruccions de codi es **repeteixen una vegada i una altra** fins que la condició ja no es compleix.

---

## 🔹 Format general d’un bucle `while`

```python
while condició:
    instruccions_de_codi
else:
    instruccions_finals


````python
x = 0

while x < 10:
    print('x és actualment:', x)
    print('x encara és més petit que 10, s’hi afegeix 1')
    x += 1
````

Observa quantes vegades s’han executat les instruccions **`print`** i com el bucle **`while`** ha continuat funcionant fins que la condició ha deixat de ser certa, és a dir, quan **`x == 10`**.  

És important remarcar que, un cop això passa, **el codi s’atura**.  

Vegem ara com podem afegir una **instrucció `else`**:

````python
x = 0

while x < 10:
    print('x és actualment:', x)
    print('x encara és més petit que 10, s’hi afegeix 1')
    x += 1

else:
    print('Hem acabat!')
````

# **break**, **continue**, **pass**

Podem utilitzar les instruccions **`break`**, **`continue`** i **`pass`** dins dels nostres bucle (**for**, i **while**) per afegir funcionalitats addicionals segons cada cas.  

Les tres instruccions es defineixen així:

- **`break`**: surt del bucle més proper en execució.  
- **`continue`**: torna a l’inici del bucle més proper (salta la resta de codi de l’iteració actual).  
- **`pass`**: no fa absolutament res (s’utilitza com a marcador de posició).

---

Si pensem en les instruccions **`break`** i **`continue`**, el format general d’un bucle **`while`** seria el següent:

```python
while condicio_0: 
    instruccions_de_codi
    if condicio_1:
        break
    if condicio_2:
        continue
else:
    instruccions_finals


````python
x = 0
print(f"x val {x}\n")
while x < 7:
    print(f'x val {x}, per tant encara és més petit que 7, s’hi afegeix 1.')
    x += 1
    print(f"Ara x val {x}.")
    if x % 2 == 0:
        print('x és parell. Seguim... \n')
        continue
    elif x == 1:
        print('x és exactament 1. Seguim... \n')
        continue
    elif x == 5:
        print(f"Quan x={5}, darrere posem un break i, per tant, sortim definitivament del bucle.\n")
        break

    print("Aquí només hauríem d’arribar si x no és parell i x no és 5.\n")

print("Ja no estem al while.")
````

Observa que l’altra instrucció **`else`** **no s’ha arribat a executar**, i que tampoc s’ha mostrat cap missatge de “continuació”!  

Després d’aquests exemples breus però senzills, ja hauries de sentir-te **còmode utilitzant instruccions `while`** en el teu codi.
