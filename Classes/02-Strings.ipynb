# Strings

Això vol dir que podrem usar *indexació* per escollir caràcters concrets dins de la sequència de caràcters.

La idea de sequència és important a Python i la tornarem a tocar amb altres tipus de variables.

## Indexació en els String Indeing
Ja hem comentat que els strings són una seqüència, això vol dir que podem usar índexos per cridar parts o elements d'aquesta. Ara aprendrem com funciona l'indexació.

A Python, usem claus <code>[]</code> després d'un objecte per cridar el seu índex. És important saber que la indexació, a Python, comença en el 0 (en altres llenguatges de programació, comença a l'1). Això vol dir que el primer caràcter d'un string estarà ubicat en la posició 0. Creem una variable que contingui un string i fem alguns exemples.

```python
# Assignar un string a la variable s
s = 'Moltes mercès'
```

```python
#Comprovem
print(s)
```

Comencem a indexar!

```python
# Primer element!
s[0]
```

```python
s[1]
```

```python
s[2]
```

Podem usar els dos punts <code>:</code> per fer un *tall de cadena* o *slicing*, el qual selecciona una secció de l'string. 

```python
# Agafem els caràcters entre la posició 1 i la posicio 4 (posició 4 no inclosa)
s[1:4]
```

```python
# Agafem tot, des de la posició 1 en endavant.
s[1:]
```

```python
# Fixa't que no hem fet cap canvi en la variable original
s
```

```python
# Agafar tot fins a l'index numero 3 (no inclòs)
s[:3]
```

Note the above slicing. Here we're telling Python to grab everything from 0 up to 3. It doesn't include the 3rd index. You'll notice this a lot in Python, where statements and are usually in the context of "up to, but not including".

```python
#Per agafar tota la cadena usem [:]. 
s[:]
```

També podem usar índexos negatius per moure'ns cap enrere. L'últim índex és l'índex -1

```python
# Agafar la última lletra
s[-1]
```

```python
# Agafar-ho tot excepte la última lletra
s[:-1]
```

També podem usar la notació [inici:final:pas] per indicar en quin index comencem, en quin index ens aturem, i quin el salt dels índexos. Quan no especifiquem el pas, aquest és 1 per defecte.

```python
# Agafar-ho tot, amb un pas de 1
s[::1]
```

```python
# Agafar-ho tot, però amb un pas de 2
s[::2]
```

```python
# Podem donar pas -1 per escriure l'string del revés
s[::-1]
```

## Mètodes inclosos en els strings

Els objectes a Python acostumen a tenir mètodes inclosos (built-in methods). Aquests mètodes són funcions que estàn adherides a l'objecte/variable i que poden fer accions sobre l'objecte/variable. 

Cridem aquests mètodes escrivint l'objecte, amb un punt, i a continuació el nom del mètode. Ens els trobarem d'aquesta forma:

objecte.mètode(paràmetres)

Els paràmetres seran arguments extra que haurem de passar als mètodes, i variaran en cada mètode. No et preocupis si ara no ho entens del tot, a continuació veurem alguns exemples i més endavant crearem els nostres objectes i els seus propis mètodes!

Alguns exemples de mètodes en strings:

```python
s
```

```python
# Escriure un string en majúscules
s.upper()
```

```python
# Escriure'l en minúscules
s.lower()
```

```python
# Separar un string pels espais en blanc (si no especifiquem res, utilitza l'espai en blanc com a separador)
s.split()
```

```python
# Separar un string per un element específic (no inclou l'element pel qual es fa la separació). 
s.split('e')
```

Hi ha molts mètodes a aplicar sobre un string! Es poden veure tots en el següent enllaç: https://docs.python.org/3/library/stdtypes.html#string-methods

# Formatejar strings

Formatejar strings et permet introduir variables dins dels strings, enlloc de concatenar els trings. Vegueu un exemple:

    jugador = 'Tomàs'
    punts = 33
    
    'Ahir pel matí, ' + jugador + ' va fer '+str(punts)+' punts.'  # concatenació d'strings
    
    f'Ahir pel matí, {jugador} va fer {punts} punts.'          # string formatejat

La manera principal que s'usa avui dia per a fer formateig d'strings és a través de **f-strings**. Però està bé que sapigueu que hi ha altres maneres, com ara usant el símbol '%' o també el mètode .format().

## F-strings

A partir de la versió de Python 3.6, es poden usar els f-strings per al formateig d'strings. Identifiquem un f-string si veiem que hi ha una **f** abans de les cometes.

```python
name = 'Leo'
edat = 35
print(f"Aquell senyor que camina per l'entrada de l'edifici es diu {name}.")
print(f"En {name} és l'encarregat del material i té {edat} anys.")
```

#### Formatejar com mostrem nombres decimals a Python 
Segueix la estructura `f"text {variable:amplada.nombre_decimals}"`

On **variable** és la variable que volem mostrar, **amplada** és el nombre de caràcters  mínims reservat per a mostrar-lo, i **nombre_decimals** és el nombre de dígits decimals que mostrarem. El nombre es mostrarà en notació exponencial.

Si al final, afegim una f, el nombre es mostrarà en decimals.

```python
num = 23.45678
print(f"Mostraré la variable num en forma exponencial, amb 10 espais reservats, i 2 decimals: {num:5.2}")
print(f"Mostraré la variable num en forma decimal, amb 10 espais reservats, i 2 decimals: {num:5.2f}")
```
