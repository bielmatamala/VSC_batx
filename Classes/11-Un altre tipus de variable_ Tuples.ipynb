# Tuples  

En Python, les **tuples** són molt semblants a les **llistes**, però amb una diferència fonamental: **són immutables**, és a dir, **no es poden modificar** després de ser creades.  

S’utilitzen per representar **informació que no hauria de canviar**, com ara els **dies de la setmana** o les **dates d’un calendari**.

---

1. Construcció de tuples  
2. Mètodes bàsics de les tuples  
3. Immutabilitat  
4. Quan utilitzar tuples  

---

A partir del que ja saps sobre les llistes, podràs tenir una **intuïció natural**  
de com fer servir les tuples, ja que es comporten de manera molt similar —  
l’única gran diferència és que **les tuples no es poden canviar**.


```python
# Crear una tupla
t = (1,2,3)
```

```python
# Obtenir la llargada de la tupla
len(t)
```

```python
# També pot tenir objectes de diferents tipus
t = ('hola',2,3,"adéu", 4,5,6)

# Veiem-lo
print(t)
```

```python
# Usar indexació de mateixa manera que amb les llistes
t[0]
```

```python
# Slicing
t[1:-1:2]
```

## Mètodes bàsics de les tuples  

Les **tuples** tenen alguns **mètodes integrats**, encara que **no tants com les llistes**.  

Vegem-ne dos dels més utilitzats:

```python
# Fes servir .index per introduir un valor i obtenir la seva posició dins la tupla
print(t.index("hola"))
print(t.index(3))
```

```python
# Utilitza .count() per comptar el nombre de vegades que apareix un valor
print(t.count(2))
```

## Immutabilitat  

Cal remarcar-ho bé: **les tuples són immutables**.  
Això vol dir que **no es poden modificar** un cop creades.  

Per deixar-ho ben clar:

```python
t[0]= 'nou text'
```

A causa d’aquesta **immutabilitat**, les tuples **no poden créixer**.  
Un cop creada una tupla, **no podem afegir-hi nous elements**.

```python
t.append('sembla que no')
```

## Quan utilitzar tuples  

Et pots preguntar: “Per què fer servir tuples si tenen menys mètodes disponibles que les llistes?”  

La veritat és que les **tuples no s’utilitzen tan sovint com les llistes**, però són molt útils **quan necessitem immutabilitat**.

Si en el teu programa passes un objecte entre funcions o mòduls i vols assegurar-te que **no es modifiqui**,  
aleshores una **tupla** és la solució adequada.  

Proporciona una manera senzilla de **mantenir la integritat de les dades**.

Ara ja hauries de saber **crear i utilitzar tuples** en els teus programes,  
així com **entendre la seva immutabilitat**.
