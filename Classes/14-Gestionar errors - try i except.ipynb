En aquesta lliçó aprendrem sobre els **errors** i la **gestió d’excepcions** en Python.  

Segurament ja t’has trobat amb alguns errors fins ara en el curs.  
Per exemple:

````python
print('Hola)
````

Observa que obtenim un **SyntaxError**, amb la descripció addicional que es tracta d’un **EOL (End of Line Error)** mentre s’analitzava la cadena de text.

Entendre els diferents **tipus d’errors** t’ajudarà a **depurar el codi molt més ràpidament**.

---

Aquest tipus d’error i la seva descripció s’anomenen **excepció** (*Exception*).  
Encara que una instrucció o expressió sigui **sintàcticament correcta**, pot provocar un error en el moment d’executar-se.  

Els errors detectats durant l’execució s’anomenen **excepcions**,  
i no sempre són errors fatals: es poden **identificar i gestionar** dins del nostre codi.

---

Ara aprendrem **com gestionar errors i excepcions** en el nostre propi codi.

## try i except  

La terminologia i la sintaxi bàsiques per **gestionar errors** en Python són les instruccions **`try`** i **`except`**.  

El codi que pot causar una excepció es col·loca dins del bloc **`try`**,  
i la **gestió de l’excepció** s’implementa dins del bloc **`except`**.  

La sintaxi general és la següent:

```python
try:
    # Aquí executes les teves operacions
    ...
except Excepcio1:
    # Si es produeix l’excepció Excepcio1, s’executa aquest bloc
except Excepcio2:
    # Si es produeix l’excepció Excepcio2, s’executa aquest bloc
    ...
else:
    # Si no es produeix cap excepció, s’executa aquest bloc

````python
try:
    nombre = int(input("Dona'm un nombre (prova de passar un text, i després un número)"))
except:
    # Si no ha pogut convertir l'string a int (enter), este
    print("No hem pogut transformar l'input a nombre enter.")
else:
    print(f"Has donat correctament el nombre n")
````

Genial! Fixa’t que **només hem imprès un missatge**, però el codi ha continuat executant-se amb normalitat.

Això és **molt útil** quan volem tenir en compte **possibles errors d’entrada o execució** dins del nostre programa.  

Aquesta tècnica ens permet **preveure errors** i continuar executant el codi,  
en lloc que el programa es trenqui completament com passava abans. 


Ara bé, què passa si volem **seguir executant codi després d’una excepció**?  
Aquí és on entra en joc el bloc **`finally`**.

---

## finally  

El bloc **`finally:`** s’executa **sempre**,  
tant si hi ha hagut una excepció dins del bloc `try` com si no.  

La seva sintaxi és la següent:

```python
try:
    # Bloc de codi aquí
    ...
    # Si es produeix una excepció, aquest codi podria ser omès
except: 
   # codi que s'executa fa en cas d'haver-hi una excepció
else:
   # codi que s'executa en cas que no tinguem una excepció
finally:
    # Aquest bloc s'executarà sempre


Un exemple complet:

````python
### Exemple complet amb `try`, `except`, `else` i `finally`
try:
    # Intentem obrir un fitxer per llegir-lo
    fitxer = open("dades.txt", "r")
    contingut = fitxer.read()
except FileNotFoundError:
    # Si el fitxer no existeix, es captura aquesta excepció
    print("❌ Error: el fitxer no s’ha trobat.")
except PermissionError:
    # Si no tenim permisos de lectura
    print("⚠️ No tens permisos per llegir aquest fitxer.")
else:
    # Si no hi ha cap error, s’executa aquest bloc
    print("✅ Fitxer llegit correctament!")
    print("Contingut:")
    print(contingut)
finally:
    # Aquest bloc s’executa sempre, passi el que passi
    try:
        fitxer.close()
        print("📁 Fitxer tancat correctament.")
    except NameError:
        # Si el fitxer mai es va obrir (perquè hi havia un error abans)
        print("ℹ️ El fitxer no s’ha arribat a obrir, res a tancar.")
````

**Genial!** 

Ara ja saps **com gestionar errors i excepcions en Python** utilitzant la notació  
**`try`**, **`except`**, **`else`** i **`finally`**.  

Aquests blocs et permetran escriure **codi més segur, controlat i professional**, evitant que el teu programa es trenqui ("peti") davant d’errors inesperats.
