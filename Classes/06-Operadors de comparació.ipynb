# Operadors de comparació  

En aquesta lliçó aprendrem sobre els **operadors de comparació** en Python.  
Aquests operadors ens permetran **comparar variables** i obtenir un valor **booleà** (`True` o `False`).  

Si tens algun tipus de formació en matemàtiques, aquests operadors t’haurien de resultar molt intuïtius.

Primer presentarem una **taula amb els operadors de comparació** i després veurem alguns exemples.

---

## Taula d’operadors de comparació  

En la taula següent, `a = 3` i `b = 4`.

<table class="table table-bordered">
<tr>
<th style="width:10%">Operador</th>
<th style="width:45%">Descripció</th>
<th>Exemple</th>
</tr>

<tr>
<td>==</td>
<td>Si els valors dels dos operands són iguals, la condició és certa.</td>
<td>(a == b) no és cert.</td>
</tr>

<tr>
<td>!=</td>
<td>Si els valors dels dos operands no són iguals, la condició és certa.</td>
<td>(a != b) és cert.</td>
</tr>

<tr>
<td>&gt;</td>
<td>Si el valor de l’operand esquerre és més gran que el del dret, la condició és certa.</td>
<td>(a &gt; b) no és cert.</td>
</tr>

<tr>
<td>&lt;</td>
<td>Si el valor de l’operand esquerre és més petit que el del dret, la condició és certa.</td>
<td>(a &lt; b) és cert.</td>
</tr>

<tr>
<td>&gt;=</td>
<td>Si el valor de l’operand esquerre és més gran o igual que el del dret, la condició és certa.</td>
<td>(a &gt;= b) no és cert.</td>
</tr>

<tr>
<td>&lt;=</td>
<td>Si el valor de l’operand esquerre és més petit o igual que el del dret, la condició és certa.</td>
<td>(a &lt;= b) és cert.</td>
</tr>
</table>

---


Anem a veure alguns exemples snzills.
#### Igual

```python
2 == 2
```

```python
1 == 0
```

Fixeu-vos que <code>==</code> és un operador de <em>comparació</em>, mentre que <code>=</code> és un operador d'<em>assignació</em>.

#### No igual

```python
2 != 1
```

```python
2 != 2
```

Podeu utilitzar el **not** per negar una cosa:

```python
not (2 == 2)
```

```python
2 > 3
```

```python
not ( 2 > 3)
```

#### Més gran que

```python
2 > 1
```

```python
2 > 4
```

#### Menor que

```python
2 < 4
```

```python
2 < 1
```

#### Més gran o igual que

```python
2 >= 2
```

```python
2 >= 1
```

#### Més petit o igual que

```python
2 <= 2
```

```python
2 <= 4
```

**Genial!** Repassa cada operador de comparació per assegurar-te que entens bé què indica cadascun d’ells.  
Esperem que hagi estat senzill i clar per a tu.
