# Llistes

A la secció d'strings vam introduir el concepte de seqüència a Python. Les llistes són la versió més genèrica d'una seqüència a Python. Les llistes, no com els strings, són objectes mutables, de manera que podem canviar els seus elements de manera individual. 

En aquesta secció aprendrem sobre:
    
    1.) Crear llistes
    2.) Indexació i agafar seccions de llistes (list slicing)
    3.) Mètodes bàsics en les llistes
    4.) Anidament de llistes (llistes dins de llistes)
    5.) Introducció a List Comprehension 

Les llistes es construeixen amb bràquets [], usant comes ',' per separar els elements dins de la llista. 

```python
# Assignar una llista a una variable anomenada la_meva_llista
la_meva_llista = [1,2,3]
```

Hem creat una llista de nombres enters ("integers"), però dins d'una llista podem tenir elements de diferent tipus. Per exemple:

```python
la_meva_llista = ['un string',23,100.232,'o']
```

Com amb els strings, la funció len() ens diu la quantitat d'elemetns dins de la seqüència. 

```python
len(la_meva_llista)
```

### Indexació i Slicing
Indexació i slicing funcionen de la mateixa manera que amb els strings. Construim un exemple:

```python
la_meva_llista = ['un','dos','tres',4,5]
```

```python
# Agafar el primer element 
la_meva_llista[0]
```

```python
# Agafar des de l'índex 1 fins al final: 
la_meva_llista[1:]
```

```python
# Agafar tots els elements fins a l'índex 3 , o reordenar la llista del revés
print(la_meva_llista[:3])
print(la_meva_llista[::-1])
```

També podem usar la suma + per concatenar llistes, com havíem fet amb els strings. 

```python
la_meva_llista + ['nou ítem']
```

Doneu-vos compte que això no canvia la llista original. 

```python
la_meva_llista
```

Hauríem de canviar-la mitjançant a una assignació usant el símbol = per canviar-la permanentment. 

```python
# Reassign
la_meva_llista += ['ítem afegit permanentment']
```

```python
la_meva_llista
```

També podem usar el símbol * per a duplicar la llista, de manera similar al que vam veure amb els strings.

```python
# Make the list double
la_meva_llista * 2
```

```python
# De nou, aquesta dupliacció no s'ha fet de manera permanent.  Hauríem de fer la_meva_llista = la_meva_llista*2
la_meva_llista
```

## Mètodes bàsics en llistes

Si estàs familiaritzat amb altres llenguatges de programació, potser veus algunes similituds entre arrays i les llistes de Python. 

Les llistes són elements més flexibles, ja que la seva mida no tenen una mida fixada, i no tenen cap tipus fixat (és a dir, poden contenir elements de diferent tipus). 
Explorem alguns mètodes de llistes:

```python
# Assignar una llista a una variable anomenada list1
list1 = [1,2,3]
```

Utilitzar el mètode append per afegir permanentment un element a la llista. 

```python
# Append
list1.append('Nou element!')
```

```python
# Show
list1
```

Usar **pop** per eliminar un element de lallista. Per defecte, s'elimina l'últim element, però podem especificar quin element volem eliminar passant el seu índex. 

```python
# Eliminar l'element amb índex 0
list1.pop(0)
```

```python
# Mostrem la llista
list1
```

```python
# Assignar a una variable nova, l'element que eliminem. Si no especifiquem l'índex, es considera l'últim element (índex -1)
popped_item = list1.pop()
```

```python
popped_item
```

```python
# Mostrar la llista resultant. 
list1
```

Cal tenir en compte que si especifiquem un índex que no existeix a la llista, Python ens retornarà un error:

```python
list1[100]
```

Podem usar la funció **sort** i el mètode **reverse** per ordenar la llista:

```python
new_list = ['a','e','x','b','c']
print(new_list)
```

```python
# Revertir l'ordre de la llista
new_list.reverse()
print(new_list)
```

```python
# Ordenar la llista
new_list.sort()
print(new_list)
```

El mètode sort reordena la pròpia llista.
Si volem assignar la llista ordenada a una nova variable, hem d'usar la funció sorted.

```python
# Usar la funció "sorted" per ordenar una llista. Seguirà l'ordre alfabètic (si són strings), i ascenedent, si són números. 
llista_ordenada = sorted(new_list)
```

```python
print(llista_ordenada)
```

## Llistes anidades 
Les estructures de dades a Python permeten el que diem "anidaments" o "nesting". Això sigifica que podem tenir estructures de dades dins d'estructures de dades (variables dins de variables). Per exemple: una llista dins d'una llista. 
Veiem'ho!

```python

lst_1=[1,2,3]
lst_2=[4,5,6]
lst_3=[7,8,9]

# Fem una llista de llistes que formen una matriu ("matrix")
matrix = [lst_1,lst_2,lst_3]
```

```python
# Show
print(matrix)
print("Podem imaginar-ho com una graella amb files i columnes, on cada fila son les llites que hem definit anteriorment ")
for el in matrix:
    print(el)
```

Podem usar l'indexació per agafar elements, però ara hi ha dos nivells d'indexació. Els elements dins de la matriu, i depsrés els elements dins de la llista! 

```python
# Agafar el segon elements dins la matriu
matrix[1]
```

```python
# Agafar alguns elements de la matriu: 
print(f"L'element de la primera fila i primera columna és: {matrix[0][0]}")
print(f"L'element de la segona fila i primera columna és: {matrix[1][0]}")
print(f"L'element de la tercera fila i segona columna és: {matrix[2][1]}")
print(f"L'element de la tercera fila i tercera columna és: {matrix[2][2]}, o també: {matrix[-1][-1]}")
```

# List Comprehension -- Creació ràpida de llistes
Python té una propietat avançada anomenada list comprehension. Això ens permet crear llistes més ràpidament, ens molts casos. Per a entendre-ho plenament, hem d'entendre bé els loops, així que és possible que ara mateix algú es quedi amb algun dubte. Tot i això, no et preocupis, ara mateix només ho avancem, i més endavant ho mirarem amb més detall.  

But in case you want to know now, here are a few examples!

```python
# Crear una llista usant list comprehension. Estem agafant el primer element de cada element dins la llista matrix. 
x = [row[0] for row in matrix]
```

Intenta visualitzar què ha de tenir la variable x abans d'executar la següent cel·la:

```python
x
```
