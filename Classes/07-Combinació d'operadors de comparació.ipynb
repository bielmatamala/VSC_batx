# Operadors de comparació en cadena
Una característica interessant de Python és la possibilitat de **concatenar o encadenar** múltiples comparacions per realitzar una prova més complexa.
Pots utilitzar aquestes comparacions en cadena com una abreviatura d’expressions booleanes més llargues.

En aquesta lliçó aprendrem com encadenar operadors de comparació i també introduirem dues instruccions molt importants en Python: 

**and** i **or**.

Vegem alguns exemples d’ús d’aquestes cadenes:

La sentència prèvia següent comprova si 1 és més petit que 2, i 2 és més petit que 3:

```python
1<2 and 2<3
```

El **and** s'usa per assegurar-se que les dues sentències que estan a banda i banda de l'**and** han de ser certes, per a que la sentència total també ho sigui. Mire'm un altre exemple:

```python
1<3 and 3>2
```

```python
2==3 and 3<4
```

És important tenir clar que el Python està comprovant les dues instàncies de les comparacions.

També podem usar l'**or** (o) per a escriure comparacions en Python. Per exemple:

```python
1==2 or 2<3
```

En aquest cas (**or**), veiem que la sentència total ha sigut certa. Això és degut a que l'operador **or** només exigeix que una de les dues sentències sigui certa (o una, o l'altra). 

Mire'm més exemples per acabar d'entendre-ho.

```python
1==1 or 100==1
```

```python
2<3 or 100>1000
```

```python
1 != 1 or 4 < 5
```

Un cop acabat aquest notebook t'hauries de sentir còmode entenent com es comporten el **and** i el **or**. Pots posar-te petits exemples per veure que es compleix el que vols veure. També pots provar combinant-ne més!!

```python
((2 < 3 ) and (3 > 4)) or (5 < 6)
```
