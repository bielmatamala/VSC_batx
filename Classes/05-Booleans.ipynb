## Booleans

Python inclou Booleans, que són un tipus de dades que només poden valdre "True" ("cert") o "False" ("fals"), que equivalent bàsicament als enters 1 i 0. Els booleans ens serviràn per fer bifurcacions en el flux dels codis a través d'expressions condicionals: si aquesta cosa passa, fes això; però si passa aquesta altra cosa, fes aquesta altra. Ho veurem més endavant quan veguem les sentències if, elif, else, i while, principalment. 
Veiem alguns exemples de Booleans.

```python
# Definim una variable que tingui un booleà
a = True
```

```python
#Show
a
```

```python
type(a)
```

També podem usar comparacions per a definir booleans. Els veurem més endavant, però per exemple fixeu-vos en les següents cel·les: 

```python
# Clarament, 1 no és més gran que 2, així que Python ens retorna False
1 > 2
```

```python
# En canvi:
2 <= 3
```

També està bé que conegueu que hi ha un tipus de dades que són els "Nonetype" que bàsicament es creen usant la comanda None. 

```python
b = None
```

```python
# Show
print(b)
print(type(b))
```
