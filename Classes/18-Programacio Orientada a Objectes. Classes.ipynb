## class
Els objectes definits per l’usuari es creen utilitzant la paraula clau <code>class</code>. Una classe és un plànol que defineix la naturalesa d’un objecte que usarem en el futur. A partir de les classes podem construir instàncies. Una instància és un objecte concret creat a partir d’una classe determinada. Per exemple, més amunt hem creat l’objecte <code>lst</code>, que era una instància d’un objecte de tipus llista.

Vegem com podem utilitzar <code>class</code>:


````python
# Crear un nou tipus d’objecte anomenat Mostra
class Mostra:
    pass

# Instància de Mostra
x = Mostra()

print(type(x))
````

Per convenció, donem a les classes un nom que comença amb majúscula. Observa com <code>x</code> és ara la referència a la nostra nova instància d’una classe Mostra. En altres paraules, **instanciem** la classe Mostra.

Dins de la classe, actualment només tenim <code>pass</code>, però podem definir atributs de classe i mètodes.

Un **atribut** és una característica d’un objecte.  
Un **mètode** és una operació que podem realitzar amb l’objecte.

Per exemple, podem crear una classe anomenada Gos. Un atribut d’un gos pot ser la seva raça o el seu nom, mentre que un mètode d’un gos pot ser un mètode <code>.borda()</code> que retorna un so.

Vegem millor què són els atributs mitjançant un exemple.

## Atributs
La sintaxi per crear un atribut és:

    self.attribute = alguna_cosa

Hi ha un mètode especial anomenat:

    __init__()

Aquest mètode s’utilitza per inicialitzar els atributs d’un objecte. Per exemple:

````python
class Gos:
    def __init__(self, raça):
        self.raça = raça
        
frodo = Gos(raça='Lab')
sam = Gos(raça='Huskie')
````

Desglossem el que hem vist més amunt. El mètode especial

    __init__()

es **crida automàticament** just després que l’objecte hagi estat creat:

    def __init__(self, raça):

Cada atribut dins de la definició d’una classe comença amb una referència a la instància de l’objecte. Per convenció, aquesta referència s’anomena `self`. `raça` és l’argument, i el seu valor es passa durant la instanciació de la classe.

    self.raça = raça


Ara hem creat dos instàncies de la classe Gos. Podem accedir a les seves raçes així:

````python
print(frodo.raça)
print(sam.raça)
````

Observa que no tenim cap parèntesi després de `raça`; això és perquè és un atribut i no rep cap argument.

En Python també existeixen els *atributs d’objecte de classe*. Aquests atributs de classe són iguals per a qualsevol instància de la classe. Per exemple, podríem crear l’atribut *especie* per a la classe Gos. Els gossos, independentment de la seva raça, nom o altres atributs, sempre seran mamífers. Apliquem aquesta lògica de la manera següent:

````python
class Gos:
    
    # Class Object Attribute
    especie = 'mamífer'
    
    def __init__(self, raça, nom):
        self.raça = raça
        self.nom = nom
````

````python
sam = Gos('Lab','Sam')
````

````python
sam.nom
````

Observa que l’atribut d’objecte de classe es defineix fora de qualsevol mètode dins de la classe. A més, per convenció, el col·loquem primer, abans del mètode `__init__` (tot i que també podríem posar-ho en el __init__; quan és independent de la instància, com en aquest cas, podem definir-ho prèviament)

````python
sam.especie
````

## Mètodes

Els mètodes són funcions definides dins del cos d’una classe. S’utilitzen per realitzar operacions amb els atributs dels nostres objectes. Els mètodes són un concepte clau del paradigma de la POO i són essencials per dividir responsabilitats en la programació, especialment en aplicacions grans.

Bàsicament, pots pensar en els mètodes com funcions que actuen sobre un objecte i que tenen en compte el mateix objecte mitjançant l’argument *self*.

Vegem un exemple de creació d’una classe Cercle:


````python
import math
class Cercle:

    # El cercle s’instancia amb un radi (per defecte és 1)
    def __init__(self, radi=1):
        self.pi = math.pi
        self.radi = radi
        self.area = radi * radi * self.pi

    # Mètode per reinicialitzar el radi
    def ActualitzaRadi(self, nou_radi):
        self.radi = nou_radi
        self.area = nou_radi * nou_radi * self.pi

    # Mètode per obtenir la circumferència
    def CalculaPerimetre(self):
        return 2 * self.radi * self.pi


c = Cercle()

print('El radi és: ', c.radi)
print('L’àrea és: ', c.area)
print('El perímetre és: ', c.CalculaPerimetre())
````

````python
c.ActualitzaRadi(2)

print('Radi és: ',c.radi)
print('Àrea és: ',c.area)
print('Perímetre és: ',c.CalculaPerimetre())
````

Genial! Observa com hem utilitzat la notació `self.` per fer referència als atributs de la classe dins de les crides als mètodes. Repassa com funciona el codi anterior i prova de crear el teu propi mètode.

## Herència

L’herència és una manera de crear noves classes a partir de classes que ja han estat definides prèviament. Les noves classes creades s’anomenen classes derivades, i les classes de les quals deriven s’anomenen classes base. Alguns beneficis importants de l’herència són la reutilització de codi i la reducció de la complexitat d’un programa. Les classes derivades (descendents) poden sobreescriure o ampliar la funcionalitat de les classes base (ancestres).

Vegem-ne un exemple incorporant el treball previ amb la classe Gos:


````python
class Animal:
    def __init__(self):
        print("Animal creat")

    def QuiSoc(self):
        print("Animal")

    def menjar(self):
        print("Menjant")


class Gos(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Gos creat")

    def QuiSoc(self):
        print("Gos")

    def bordar(self):
        print("Bup bup!")
````

````python
g = Gos()
````

````python
g.QuiSoc()
````

````python
g.menjar()
````

````python
g.bordar()
````

En aquest exemple, tenim dues classes: Animal i Dog. Animal és la classe base i Dog és la classe derivada.

La classe derivada hereta la funcionalitat de la classe base.

* Això es mostra amb el mètode `eat()`.

La classe derivada modifica el comportament existent de la classe base.

* Això es mostra amb el mètode `whoAmI()`.

Finalment, la classe derivada amplia la funcionalitat de la classe base definint un nou mètode `bark()`.

## Polimorfisme

Hem après que, tot i que les funcions poden rebre arguments diferents, els mètodes pertanyen als objectes sobre els quals actuen. En Python, el *polimorfisme* fa referència a la manera com diferents classes d’objectes poden compartir el mateix nom de mètode, i aquests mètodes poden ser cridats des del mateix lloc encara que s’hi passin objectes diferents. La millor manera d’explicar-ho és amb un exemple:


````python
class Gos:
    def __init__(self, nom):
        self.nom = nom

    def parlar(self):
        return self.nom + ' et diu Bup!'
    
class Gat:
    def __init__(self, nom):
        self.nom = nom

    def parlar(self):
        return self.nom + ' et diu Miau!' 
    
coco = Gos('Coco')
mysha = Gat('Mysha')

print(coco.parlar())
print(mysha.parlar())
````

Aquí tenim una classe Dog i una classe Cat, i cadascuna té un mètode `.parlar()`. Quan es crida, el mètode `.parlar()` de cada objecte retorna un resultat únic per a aquell objecte.

Hi ha diverses maneres de demostrar el polimorfisme. Primer, amb un bucle `for`:

````python
for mascota in [coco,mysha]:
    print(mascota.parlar())
````

Una altra vista és amb funcions:

````python
def mascota_parlar(mascota):
    print(mascota.parlar())

mascota_parlar(coco)
mascota_parlar(mysha)
````

En ambdós casos hem pogut passar diferents tipus d’objectes i hem obtingut resultats específics de cada objecte mitjançant el mateix mecanisme.

Una pràctica més habitual és utilitzar classes abstractes i herència. Una classe abstracta és aquella que no s’espera que sigui instanciada mai. Per exemple, mai tindrem un objecte Animal, sinó només objectes Gos i Gat, tot i que els gossos i els gats deriven de la classe Animal:

````python
class Animal:
    def __init__(self, nom):    # Constructor de la classe
        self.nom = nom

    def parlar(self):              # Mètode abstracte que definim per convenció
        raise NotImplementedError("Subclass must implement abstract method")


class Gos(Animal):
    
    def speak(self):
        return self.nom+' diu Bup!'
    
class Gat(Animal):

    def speak(self):
        return self.nom+' diu Miau!'
    
fido = Gos('Fido')
isis = Gat('Isis')

print(fido.speak())
print(isis.speak())
````

Exemples de polimorfisme en la vida real inclouen:
* obrir diferents tipus de fitxers — es necessiten eines diferents per visualitzar fitxers Word, PDF i Excel
* sumar diferents objectes — l’operador `+` realitza tant operacions aritmètiques com concatenació

## Mètodes especials

Finalment, repassem els mètodes especials. Les classes en Python poden implementar certes operacions mitjançant noms de mètodes especials. Aquests mètodes no es criden directament, sinó que Python els invoca mitjançant una sintaxi específica del llenguatge. Per exemple, creem una classe Llibre:

````python
class Llibre:
    def __init__(self, titol, autor, pagines):
        print("S'ha creat un llibre")
        self.titol = titol
        self.autor = autor
        self.pagines = pagines

    def __str__(self):
        return "Títol: %s, autor: %s, pàgines: %s" % (self.titol, self.autor, self.pagines)

    def __len__(self):
        return self.pagines

    def __del__(self):
        print("S'ha destruït un llibre")
````

````python
llibre = Llibre("El Python ho peta!", "David Araujo", 159)

#Special Methods
print(llibre)
print(len(llibre))
del llibre
````

Aquests mètodes especials es defineixen mitjançant l’ús de guions baixos. Ens permeten utilitzar funcions específiques de Python sobre objectes creats a partir de la nostra classe.

## Exemples de propietats @property

En Python, una **propietat** és un mètode que es calcula com una funció,
però s’utilitza com si fos un atribut (sense parèntesis).

**Exemple amb la diagonal d'un quadrat.**

````python
import math

class Rectangle:
    def __init__(self, ample, alt):
        self.ample = ample
        self.alt = alt

    def area(self):
        return self.ample * self.alt

    @property
    def diagonal(self):
        return math.sqrt(self.ample**2 + self.alt**2)
````

````python
r = Rectangle(3, 4)
print(r.area())      # metode
print(r.diagonal)    # propietat (sense parentesis)
````

La diagonal es calcula automàticament a partir de l’ample i l’alt, no és necessari cap input extern.

## Propietats amb control de valors (getter + setter)

Les propietats també permeten controlar els valors que assignem als atributs.


````python
class Quadrat:
    def __init__(self, costat):
        self.costat = costat

    @property
    def costat(self):
        return self._costat

    @costat.setter
    def costat(self, valor):
        if valor <= 0:
            raise ValueError("El costat ha de ser positiu")
        self._costat = valor

    @property
    def area(self):
        return self._costat ** 2
````

````python
q = Quadrat(5)
print(q.area)
````

````python
q.costat = 10
print(q.area)
````

````python
q.costat = -3        # error
````

Penseu que el mètode **__init__** només s'executa una vegada. Nosaltres podríem sobreescriure un atribut de manera manual (com s'està fent a l'exemple anterior). Amb el @property i el @atribut.setter podem controlar que els atributs estiguin ben definits sempre.
