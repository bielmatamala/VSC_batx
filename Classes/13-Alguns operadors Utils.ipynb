## range  

La funció **`range`** et permet *generar ràpidament* una llista de nombres enters.  
Aquesta funció és molt pràctica i s’utilitza sovint, així que val la pena entendre bé com funciona.  

Accepta **tres paràmetres** opcionals:  
- **inici** (`start`)  
- **final** (`stop`)  
- **pas** (`step`)  

Vegem alguns exemples:

`````python
range(0,11)
`````

Tingues en compte que **`range`** és una **funció generadora** (*generator function*).  
Això vol dir que, per obtenir-ne realment una **llista**, cal convertir-la amb **`list()`**.  

Per exemple:
```python
list(range(0, 5))

`````python
# Fixa't que l'11 no estarà inclòs, igual que amb la
# indexació en llistes
list(range(0,11))
`````

`````python
list(range(0,12))
`````

`````python
# El tercer paràmetre és la mida del pas (step size)!
# La mida del pas indica de quant en quant saltem
# des del nombre inicial fins al següent.

list(range(0, 11, 2))
`````

`````python
list(range(0,101,10))
`````

## enumerate  

La funció **`enumerate`** és molt útil quan treballem amb **bucles `for`**.  

Imaginem la situació següent:

`````python
comptador = 0

for lletra in 'abcdefghijklmnopqrstuvwxyz':
    print(f"At index {comptador} the letter is {lletra}")
    comptador += 1
`````

Fer un seguiment de **quantes vegades ha passat el bucle** és tan habitual, que es va crear la funció **`enumerate`** precisament per **no haver-te de preocupar** de crear i actualitzar manualment una variable com `index_count` o `loop_count`.


`````python
# En aquest for, estem fent unpacking de tuples

for i,lletra in enumerate('abcde'):
    print(f"A l'índex {i} la lletra és {lletra}")

print("\nUn altre exemple:\n")

import math
for i, element in enumerate([22,-1,-4,math.pi]):
    print(f"A l'índex {i} el número és {element}")
`````

## zip  

Abans de veure com funciona **`zip`**, fixa’t en el format que realment retorna **`enumerate`**.  

Fem una petita prova convertint el seu resultat a una **llista** amb `list()`:

`````python
list(enumerate('abcde'))
`````

Era una **llista de tuples**, cosa que significa que podem fer servir el **desempaquetament de tuples** dins del nostre bucle `for`.  

Aquesta estructura de dades és molt comuna en Python, especialment quan treballem amb **llibreries externes**.  

La funció **`zip()`** ens permet crear ràpidament una **llista de tuples** "ajuntant" (*zipping*) dues llistes entre si.


`````python
mylist1 = [1,2,3,4,5]
mylist2 = ['a','b','c','d','e']
`````

`````python
# Aquesta funció també és un generador!
# Ja ho explicarem més endavant, però de moment
# convertim el resultat en una llista per veure què retorna.

zip(mylist1, mylist2)
`````

`````python
list(zip(mylist1,mylist2))
`````

Per utilitzar aquest **generador**, simplement podem fer servir un **bucle `for`**:


`````python
for item1, item2 in zip(mylist1,mylist2):
    print('Per a aquesta tupla, el primer element era {} i el segon element era {}'.format(item1, item2))
`````

## Operador **in**  

Ja hem vist la paraula clau **`in`** durant l’ús dels bucles `for`,  però també la podem utilitzar per **comprovar ràpidament si un objecte es troba dins d’una llista**.

`````python
'x' in ['x','y','z']
`````

`````python
'x' in [1,2,3]
`````

`````python
a = [1,[1,2],3]
print(1 in a)
print(2 in a)
print([1,2] in a)
print(2 in a[1])
`````

## Operador **not in**  

Podem combinar **`in`** amb l’operador **`not`** per **comprovar si un objecte o variable no és present** dins d’una llista.

`````python
'x' not in ['x','y','z']
`````

`````python
'x' not in [1,2,3]
`````

També podem negar tota la expressió:

`````python
not ('x' in [1,2,3])
`````

Ja que recordem que el not es pot usar com un operador lògic sobre booleans, que canvia:
- True --> False
- False --> True.

`````python
print(2 > 3)
print(not (2 > 3))
`````

## random  

Python inclou una **biblioteca integrada anomenada `random`**.  
Aquesta biblioteca conté moltes funcions relacionades amb nombres aleatoris, però ara en veurem només **dues de les més útils**.

`````python
from random import shuffle
`````

`````python
# Aquesta funció barreja (shuffle) la llista "in-place",
# és a dir, no retorna cap valor nou.
# En comptes d'això, modifica directament la llista original passada com a argument.

shuffle(mylist)
`````

`````python
mylist
`````

`````python
from random import randint
`````

`````python
# Retorna un nombre enter aleatori dins l’interval [a, b],
# incloent-hi tots dos extrems.

randint(0, 100)
`````

`````python
# Retorna un nombre enter aleatori dins l’interval [a, b],
# incloent-hi tots dos extrems.

randint(0, 100)
`````

## La funció **input()** en Python  

La funció **`input()`** s’utilitza per **demanar dades a l’usuari** a través del teclat.  
Quan s’executa, el programa **s’atura** i espera que l’usuari escrigui alguna cosa i preme **Enter**.  

La informació introduïda per l’usuari sempre es **guarda com una cadena de text (string)**,  
així que si vols treballar amb nombres, cal **convertir-la** amb `int()` o `float()`.

---

### 🧩 Exemple 1: demanar un nom
```python
nom = input("Com et dius? ")
print("Hola,", nom)
```
### 🧩 Exemple 2: demanar un nombre i convertir-lo
```python
edat = int(input("Quants anys tens? "))
print(f"D’aquí a 10 anys tindràs {edat + 10} anys.")
````
