# Instruccions **if**, **elif**, **else**

Les instruccions **`if`** en Python ens permeten indicar a l’ordinador que realitzi **accions alternatives** basades en un determinat conjunt de resultats.

De manera verbal, podem imaginar que li diem a l’ordinador:

> “Ei, si es dona aquest cas, executa una acció.”

Després podem ampliar aquesta idea amb les instruccions **`elif`** i **`else`**, que ens permeten dir-li a l’ordinador:

> “Ei, si es dona aquest cas, executa una acció.  
> Si no, si es dona un altre cas, executa una altra acció.  
> I si **cap** dels casos anteriors s’ha complert, executa aquesta altra acció.”

---

## 🔹 Sintaxi bàsica

Vegem ara la **sintaxi bàsica** d’una estructura `if` per entendre-ho millor:

```python
if cas1:
    executa_acció1
elif cas2:
    executa_acció2
else:
    executa_acció3


## Primer exemple

Veguem un exemple ràpid amb un **if**:

````python
if True:
    print('És cert i per aquest motiu veus aquest text!')
````

Let's add in some else logic:

````python
x = False

if x:
    print('x era cert!')
else:
    print('Això es mostrarà en cas que x no sigui cert!')
````

### Branques múltiples

Vegem una visió més completa de fins a quin punt poden portar-nos les instruccions **`if`**, **`elif`** i **`else`**!

Escriurem això en una **estructura tabulada**.  
Fixa’t bé en com les línies amb **`if`**, **`elif`** i **`else`** **s’alineen** dins del codi.  
Això t’ajudarà a veure **quin `if` està relacionat amb quin `elif` o `else`**, i a entendre millor l’estructura lògica.

També tornarem a introduir la **sintaxi de comparació** en Python.

````python
### Exemple de branques múltiples

loc = 'Banc'

if loc == 'Taller':
    print('Benvingut/da al taller!')
elif loc == 'Banc':
    print('Benvingut/da al banc!')
else:
    print('On ets?')
````

Observa com les instruccions **`if`** anidades es comproven una per una fins que alguna retorna un valor **`True`**, fet que provoca que s’executi el bloc de codi corresponent.  

També has de tenir en compte que pots utilitzar **tants `elif` com vulguis** abans d’acabar l’estructura amb un **`else`**.

Crearem ara **dos exemples senzills més** per practicar amb les instruccions **`if`**, **`elif`** i **`else`**:


````python
### Exemple 1

persona = 'Sara'

if persona == 'Sara':
    print('Benvinguda, Sara!')
else:
    print('Benvingut/da! Com et dius?')
````

````python
persona = 'Joan'

if persona == 'Sara':
    print('Benvinguda Sara!')
elif persona == 'Joan':
    print("Benvingut Joan!")
else:
    print('Benvingut/da! Com et dius?')
````

````python
persona = 'Miquel'

if persona == 'Sara':
    print('Benvinguda Sara!')
elif persona == 'Joan':
    print("Benvingut Joan!")
else:
    print('Benvingut/da! Com et dius?')
````

# Exemple: 
Fes un programa que llegeix un nombre parell, i usant un **if**-**else** et diu si aquest nombre és parell o senar.

````python
n = int(input("Dona'm n"))
if n % 2 == 0:
    print(f"{n} és parell")
else:
    print(f"{n} és senar")
````
