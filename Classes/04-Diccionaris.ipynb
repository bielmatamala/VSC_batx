# Diccionaris

Hem vist que els strings i les llistes són seqüències a Python. Ara, canviarem una mica el punt de vista i aprendrem sobre **mappings** a Python. Els mappings fan referència a quan tenim un element identificat a través d'una etiqueta (o clau). 

Aquesta secció serà una introducció als diccionaris, i veurem: 

    1.) Construir un diccionari (Dictionary) 
    2.) Accedir als elements o objectes d'un diccionari
    3.) Diccionaris anidats (Nesting Dictionaries)
    4.) Mètodes bàsics dels diccionaris 

Com dèiem, què són els mappings? Els mappings són una colecció d'objectes que estàn emmagatzemats amb una clau, a diferència d'una seqüència que té objectes emmagatzemats només a través del seu índex. És una diferència força important, ja que dins dels diccionaris no hi ha el concepte de **ordre** implícit en l'objecte (és a dir, no podem acceidr als elements a través d'un índex o un ordre establert).

Un diccionari de Python consisteix en un conjunt de claus i els seus objectes associats. La clau ha de ser un string. El valor pot ser qualsevol objecte de Python.  


## Construir un diccionari
Veiem com es construeixen amb alguns exemples: 

```python
# Crear un diccionari amb {clau1:valor1, clau2:valor2}
my_dict = {'key1':'value1','key2':'value2'}
```

```python
# Cridar els valors a través de la seva clau.
my_dict['key2']
```

És important saber que els diccionaris de Python són elements molt flexibles que permeten guardar informació molt diferent: 

```python
diccionari_escola= {'tipus':"escola pública", 'estudiants':1000, "profesors": ["Joan", "Maria"], 'estudis': {"ESO": "SI", "BATXILLERAT": "NO"}}
```

```python
# Cridem l'element associat a la clau "estudis"
diccionari_escola['estudis']
```

```python
# Podem criar un element dins d'aquest diccionari
diccionari_escola['estudis']["ESO"]
```

```python
# Aquest últim element és una variable string. Podem utilitzar qualsevol mètode que s'utilitzi sobre un string. Per exemple,
diccionari_escola['estudis']["ESO"].lower()
```

Podem canviar directament els elements dins d'un diccionari. 

```python
diccionari_escola['estudiants']
```

```python
# Restar 123 del valor
diccionari_escola['estudiants'] = diccionari_escola['estudiants'] - 123
```

```python
# Comprovar
diccionari_escola['estudiants']
```

Com ja havíem comentat anteriorment, recordem que Python té un mètode inclòs per a restar/sumar/restar/dividir automàticament sobre una mateixa variable. Per exemple: 

```python
# Set the object equal to itself minus 123 
diccionari_escola['estudiants'] -= 123
print(diccionari_escola["estudiants"])
diccionari_escola['estudiants'] *= 2
print(diccionari_escola["estudiants"])
```

També podem crear parells clau-valor mitjançant l'assignament amb un =.  De cop, pensem que el diccionari_escola pot tenir també la clau "telèfon"

```python
diccionari_escola["telefon"] = "637101021"
print(diccionari_escola)
```

```python
# També podem crear un diccioanri buit:
d = {}
```

```python
#  I donar alguns valors:
d['nom_base_dades'] = "balanç_economic_2025"
d['password'] = "love_barça"
```

```python
# Mostrem
d
```

## Nesting with Dictionaries

Hopefully you're starting to see how powerful Python is with its flexibility of nesting objects and calling methods on them. Let's see a dictionary nested inside a dictionary:

```python
# Dictionary nested inside a dictionary nested inside a dictionary
d = {'key1':{'nestkey':{'subnestkey':'value'}}}
```

Wow! That's a quite the inception of dictionaries! Let's see how we can grab that value:

```python
# Keep calling the keys
d['key1']['nestkey']['subnestkey']
```

## Alguns mètodes dels diccionaris

```python
# Creem un diccionari genèric
d = {'key1':1,'key2':2,'key3':3}
```

```python
# Retornar totes les claus del diccionari 
d.keys()
```

```python
# Retornar els valors del diccionari
d.values()
```

```python
# Retornar tuples amb totes les parelles clau-valor. Aprendem aviat que són les tuples. 
d.items()
```
