## Tipus de nombres

Python té diferents tipus de nombres. Bàsicament ens centrarem en nombres enters ("integers") i en nombres decimals ("floating point numbers" o "floats").

Els "integers" són nombres enters, positius o negatius. Per exemple, 2 i -3 són exemples de integers. 

Els floats a Python s'identifiquen perquè apareixen amb un "." que indica el decimal; o bé perquè s'usa l'exponencial "E" per a definir el nombre. Per exemple, 2.0 i -3.1 són floats; així com també ho és 4E2 (4 * 10^2= 400).

En aquest curs principalment treballarem amb enters o floats simples (sense l'exponencial).

### Aritmètica Bàsica

```python
# Suma a+b
2+1
```

```python
# Resta a-b
2-1
```

```python
# Multiplicació a*b
2*2
```

```python
# Divisió a/b
3/2
```

```python
# Divisió entera a//b
7//4
```

**Ei! Què ha succeït? Fins on sabem, 7/4 és 1.75, no 1!**

El motiu d'obtenir aquest resultat és que estem usant la divisió entera. L'operador // trunca el decimal sense arrodonir, i retorna l'enter resultant. És a dir, retorna el *quocient* de la divisió.

**Per obtenir el residu de la divisió...**

```python
# Residu a%b
7%4
```

```python
# Potències a**b
2**3
```

```python
# També podem fer arrels quadrades elevant a 0.5
4**0.5
```

```python
# La jerarquia d'operacions típica es manté a Python
2 + 10 * 10 + 3
```

```python
# Però també podem usar parèntesis per especificar l'ordre que volem en les operacions
(2+10) * (10+3)
```

## Assignació de variables

Ara que ja hem vist com usar els nombres en Python com una calculadora, veiem com podem donar-li nom a aquests nombres i utilitzar-los com a variables. 

Utilitzarme el signe igual (=) per assignar valors a variables. Veiem alguns exemples. 

```python
# La variable "a" té valor 5
a = 5
```

D'ara en endavant, si utilitzem la variable "a", tindrà valor 5.

```python
# Suma de variables
a+a
```

Podem **reassignar** el valor de la variable. 

```python
# Reassignment
a = 10
```

```python
# Comprovació
a
```

També podem reassignar el valor usant la pròpia variable (ens serà útil més endavant)

```python
# Comprovem el valor d'a
a
```

```python
# Utilitzem a per redefinir a
a = a + a
```

```python
# Comprovem 
a
```

Els noms de les variables han de seguir algunes normes:

    1. No poden començar per un nombre.
    2. No pot haver-hi espais dins dels noms, i generalment s'usa "_" . Per exemple, si volem anomenar una variable com "capacitat bateria", llavors la definiriem com "capacitat_bateria". 
    3. No es poden usar els símbols :'",<>/?|\()!@#$%^&*~-+
    4. És considerat una bona pràctica (PEP8) que els noms estiguin en minúscules.
    5. No usar els noms "l", "O" o "I". 
    6. No usar noms reservats a funcions de python, com ara "list", "str", "int",... 


Usar variables és un principi bàsic de tots els llenguatges de programació. Utilitzar noms que donin informació respecte la variable ens serà útil i facilitarà la lectura, tant la nostra com la d'un tercer que miri el codi que hem desenvolupat.

```python
guany = 100

perdua = 75

resultat_exercici = guany - perdua
```

```python
# Mostrar el resultat de l'exercici
resultat_exercici
```

Podem comprovar el tipus d'una variable mitjançant la funció "type"

```python
type(resultat_exercici)
```
